```python
import pandas as pd
import os
import matplotlib.pyplot as plt

os.chdir("..")
```

```python
prices_df = pd.read_csv('data_/prices.csv')
features_df = pd.read_csv('data_/features.csv')
returns_df = pd.read_csv('data_/returns.csv')
```

```python
def plot_timeseries(df : pd.DataFrame):
    df = (
        df
        .assign(
            date = lambda x: pd.to_datetime(x["date"], errors="coerce")
        )
        .sort_values("date")
        .set_index("date")
    )

    ax = df.plot(figsize=(10,5))
    ax.legend(bbox_to_anchor=(1, 1))
    plt.show()
```

```python
plot_timeseries(prices_df)
```

```python
plot_timeseries(returns_df)
```
