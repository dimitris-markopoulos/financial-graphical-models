```python
import pandas as pd
import os
import matplotlib.pyplot as plt

os.chdir("..")
```

```python
df = pd.read_csv('data_/prices.csv')
```

```python
df = (
    df
    .assign(
        date = lambda x: pd.to_datetime(x["date"], errors="coerce")
    )
    .sort_values("date")
    .set_index("date")
 )
```

```python
ax = df.plot(figsize=(10,5))
ax.legend(bbox_to_anchor=(1, 1))
plt.show()
```
